```python
#importing speech_recognition module 
import speech_recognition as sr
import pyttsx3
from datetime import datetime
import platform, os, subprocess, re
from googlesearch import search
import webbrowser
from pynput.keyboard import Key,Controller
import time


#flushing previous output
import sys
sys.stdout.flush()

def openUrl(website):
    engine.say("Opening "+website)
    #engine.runAndWait()
    new=2	
    url="http://www."+website+".com"
    # Open URL in new browser window
    webbrowser.open(url,new=new) # opens in default browser 
    
def searchGoogle(query):
    engine.say("Searching "+ query + " on google...")
    url = "https://www.google.com.tr/search?q={}".format(query)
    webbrowser.open(url)
    for j in search(query, tld="com", num=10, stop=1, pause=2):
        print(j)
        
def run_command(cmd):
    engine.say("Executing "+cmd+" command. The output is: ")
    #cmd_out=subprocess.call(cmd)
    cmd_out=subprocess.check_output(cmd)
    print (cmd_out)
    #engine.say(cmd_out)
    
def file_operation(loc,file_name,oper):
    if oper=='open':
        print("Enter opening mode: r/w/a/r+/w+/a+ ")
        engine.say("Enter opening mode: r/w/a/r+/w+/a+")
        mode=input()
        f=open(loc+file_name,mode)
    elif oper=='run':
        print('running file +file_name')
        
        

#Initializing all engine parameters

engine=pyttsx3.init()
engine.setProperty('rate',150)
engine.setProperty('voice', 'English+m3')


#initializing the recognizer
r=sr.Recognizer()

#initializing the Key Controller
keyboard=Controller()

#The personal assistant asks
print("Hey,This is Miller, May I know your name please?")
engine.say("Hey, This is Miller, May I know your name please?")
engine.runAndWait()

#recording the audio from microphone
with sr.Microphone() as source:
    r.adjust_for_ambient_noise(source)
    first_voice=r.listen(source)

try:
    nm=r.recognize_google(first_voice)
    name=(str(nm)).replace('u','')
        
except sr.UnknownValueError:            #Catching error if could not understand the speech
    name="dear"

except sr.RequestError as e:            #Catching error if could not reach to google servive
    print("Could not request results from Google speech recognition service {0} : ".format(e))
    engine.say("Could not request results from Google speech recognition service {0} : ".format(e))
        

print("Hello "+ name+ ", How can I help you")
engine.say("Hello " + name+ ", How can I help you?")
engine.runAndWait()


while True:
    print ("Say something...")

    #recording the audio from the given file
    with sr.Microphone() as source:
        r.adjust_for_ambient_noise(source)
        audio=r.listen(source)
      
    print("               ...Recording completed")
 
    #Calling the google speech recognition service
    try:
        speech=r.recognize_google(audio)
    
    except sr.UnknownValueError:            #Catching error if could not understand the speech
       # engine.say("Sorry I did not get you. Please say it again.")
        continue

    except sr.RequestError as e:            #Catching error if could not reach to google servive
        print("Could not request results from Google speech recognition service {0} : ".format(e))
        engine.say("Could not request results from Google speech recognition service {0} : ".format(e))
    

    print("You said: ",speech)
    engine.say("You said " + speech)
    words=list(speech.split(' '))
    
    Bye=['bye','buy','by','shut up','bimola']
    how_ques=['how are you','how do you do','what are you doing','whats up', 'WhatsApp']
    
    good_thoughts=['I love you', 'I miss you', 'I like you']
    compliments=['intelligent', 'sweet', 'cool', 'nice','good','sexy','cute']
    bad_thoughts=['go to hell','get lost', 'I hate you', 'shut up']
    bad_comments=['stupid','poor','bad','slow','annoying','irritating','boring']
    owner=['your boss', 'your creator', 'your owner', 'create']
    okay= ['alright', 'okay','listening']   
    Run=['run','execute']
    
    if any(i in speech for i in Bye):
        print("Bye  " + name + " Nice to talk to you.")
        engine.say("Bye  " + name + " Nice to talk to you.")
        break
        
    elif 'who are you' in speech:
        engine.say("Hi " +name + " I am Miller  ")
        
    elif any(i in speech for i in owner):
        if 'who' in speech:
            engine.say(" Shweta has created me. ")
            
    elif any(i in speech for i in okay):
        if 'are you' in speech:
            engine.say(" Yes, I am listening. Go ahead. ")            
            
    elif any(i in speech for i in how_ques):
        engine.say("I am doing good. How about you? ")
        
    elif any(i in speech for i in good_thoughts):
            engine.say("I am glad to here that.")
            
    elif any(i in speech for i in bad_thoughts):
        engine.say("Ok Bye " +name+ " See you some other time. ")
        break
            
    elif 'you are' in speech:
        if any(i in speech for i in compliments):
            engine.say("Thanks for the compliment. "+name+ "  That is so nice of you. ")
    
        elif any(i in speech for i in bad_comments):
            engine.say("Aauuhh.. Sorry for the inconvenience,  I will try better next time. ")
    

    elif 'file' in words:
        verb_index=(words.index('file'))-1
        oper=words[verb_index]
        file_name=words[verb_index+2]
        location=input("Give me location")
        file_operation(location,file_name,oper)
        
    
    elif 'open' in words:
        verb_index=words.index('open')
        keyword=words[verb_index+1]
        openUrl(keyword)
    
    elif 'search' in words:
        verb_index=words.index('search')
        keyword_uni=(str(words[verb_index+1:]))
        keyword=keyword_uni.replace('u','')
        searchGoogle(keyword)
        
    elif any(i in words for i in Run):
        if 'command' in words:
            cmd_index=words.index('command')
            keyword=words[cmd_index-1]
            run_command(keyword)
            
    elif 'type' in speech:
        text=speech.replace('type','')
        keyboard.type(text)
        
                

```
